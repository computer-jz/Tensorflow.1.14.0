```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
```

#### 输入数据转换为TFRecord 模式

```python
#生成整数类型的属性
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

#生成字符串型的属性
def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

mnist=input_data.read_data_sets('/Users/jiazhuo/PycharmProjects/tensorflow1.14.0/6-CNN/MNIST_data',
                                dtype=tf.uint8,one_hot=True)

images=mnist.train.images
print("images:",images)
print("images shape",images.shape)

#训练数据所对应的标签，可以作为一个属性保存在TFRecord中
lables=mnist.train.labels
print("lables:",lables)

#训练数据的图像分辨率，可以作为example中的一个属性
pixels=images.shape[1]
print("pixels:",pixels)

num_examples=mnist.train.num_examples

#输出TFRecord文件的地址
file_name='./TFRecord/output.tfrecords'

#创建一个writer来写TFRecord文件
writer=tf.python_io.TFRecordWriter(file_name)

for index in range(num_examples):
    # 将图像矩阵转换为字符串
    image_raw=images[index].tostring()
    print("image_raw:",image_raw)
    #将一个例子转换为example pb格式，并将所有的数据信息写入到这个结构
    example=tf.train.Example(features=tf.train.Features(
        feature={
            'pixels':_int64_feature(pixels),
            'lables':_int64_feature(np.argmax(lables[index])),
            'image_raw':_bytes_feature(image_raw)
        }
    ))

    writer.write(example.SerializeToString())
writer.close()
```

#### 读取TFRecord中的数据



```python
import tensorflow as tf

#创建一个reader来读取TFRecord中文件的样例
reader=tf.TFRecordReader()

#创建一个队列来维护输入文件列表
filename_queue=tf.train.string_input_producer(['./TFRecord/output.tfrecords'])

#从文件中读出一个样例，也可以使用read_up_to函数一次性读取多个样例
_,serialized_example=reader.read(filename_queue)

#解析读入的一个样例，若需要解析多个样例，则可以使用 parse_example函数
features=tf.parse_single_example(serialized_example,
                                 features={
                                     'image_raw':tf.FixedLenFeature([],tf.string),
                                     'pixels':tf.FixedLenFeature([],tf.int64),
                                     'lable':tf.FixedLenFeature([],tf.int64)
                                 })

#tf.decode_raw 可以将字符串解析成图像对应的像素数组
image=tf.decode_raw(features['image_raw'],tf.uint8)
lable=tf.cast(features['lable'],tf.int32)
pixels=tf.cast(features['pixels'],tf.int32)

sess=tf.Session()

#多线程处理输入数据
coord=tf.train.Coordinator()
threads=tf.train.start_queue_runners(sess=sess,coord=coord)

#每次运行读取TFRcord中的一个样例，所有样例读取完成之后，在此样例中程序会从头读取
for i in range(10):
    print(sess.run([image,lable,pixels]))


```
